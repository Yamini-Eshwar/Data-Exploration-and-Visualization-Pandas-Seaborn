```python
#Introduction to seaborn #seabirn is powerful visualization (STATISTICAL VISUALIZATION) pkg in python

import matplotlib.pyplot as plt # visualization
import seaborn as sns #distribution visualization

%matplotlib inline
plt.rcParams['figure.figsize'] = 4,4
```

```python
#distributions
sns.histplot(stats["InternetUsers"], kde=True)
plt.show()
```

```python
#distributions
sns.histplot(stats["InternetUsers"], kde=True, bins=10)
plt.show()
```

```python
#Box plots
sns.boxplot(data= stats, x ="IncomeGroup", y='BirthRate', hue="IncomeGroup")
plt.show()
```

```python
sns.lmplot(data= stats, x='InternetUsers', y= 'BirthRate')
plt.show()
```

```python
sns.lmplot(data= stats, x='InternetUsers', y= 'BirthRate', fit_reg=False, hue= 'IncomeGroup') #hue for color
plt.show()
```

```python
sns.lmplot(data= stats, x='InternetUsers', y= 'BirthRate', fit_reg=False, hue= 'IncomeGroup', height=10) #hue for color
plt.show()
```
